### Puxar base de dados

```python
import pandas as pd
exportacoes_brasil_franca = pd.read_csv(r'Base de Dados - Mentoria Python - Análise de Dados\exportacoes_franca.csv').query("`Economic Block` == 'Europe'")
display(exportacoes_brasil_franca)
```

### Informações Gerais

```python
exportacoes_brasil_franca.info()
```

### Como foi a evolução das exportações para a frança ao longo dos anos em dólar?

```python
def formatar_valor(valor):
    return f'${valor:,.2f}'

exportacao_ano_df = exportacoes_brasil_franca[['Year','US$ FOB']].groupby('Year').sum()

evolucao_exp = 1 - (exportacao_ano_df.loc[2020]['US$ FOB'] / exportacao_ano_df.loc[2016]['US$ FOB'])
print(f'De 2016 para 2020 as exportações para a França cairam em: {evolucao_exp:.2%}.')

exportacao_ano_df['US$ FOB'] = exportacao_ano_df['US$ FOB'].map(formatar_valor)
display(exportacao_ano_df)
```

### Quais os produtos mais exportados (US$) ao longo de todo o período?

```python
produtos_exportados_df = exportacoes_brasil_franca[['SH4 Description','US$ FOB']].groupby('SH4 Description').sum().sort_values('US$ FOB', ascending=False)

print(f'O produto mais exportado para a França foi: "{produtos_exportados_df.index[0]}" - US$ {produtos_exportados_df.iloc[0]["US$ FOB"]:,.2f}')

produtos_exportados_df['US$ FOB'] = produtos_exportados_df['US$ FOB'].map(formatar_valor)
display(produtos_exportados_df)
```

### Em 2020 qual cidade mais exportou (US$) para a França?

```python
cidade_exportacao = exportacoes_brasil_franca.query("`Year` == 2020")[['City', 'US$ FOB']].groupby('City').sum().sort_values('US$ FOB', ascending=False)

print(f'A cidade que mais exportou para a França foi: "{cidade_exportacao.index[0]}" - US$ {cidade_exportacao.iloc[0]["US$ FOB"]:,.2f}')

cidade_exportacao['US$ FOB'] = cidade_exportacao['US$ FOB'].map(formatar_valor)
display(cidade_exportacao)
```

### O que as 2 maiores cidades exportaram tanto em 2020?

```python
exportacoes_franca_cidades = exportacoes_brasil_franca.query("`Year` == 2020 and (`City` == 'Duque de Caxias - RJ' or `City` == 'Luís Eduardo Magalhães - BA')")[['City', 'SH4 Description', 'US$ FOB']].groupby(['City', 'SH4 Description']).sum().sort_values('US$ FOB', ascending=False)

print(f'{exportacoes_franca_cidades.index[0]} - US$ {exportacoes_franca_cidades.iloc[0]["US$ FOB"]:,.2f}')
print(f'{exportacoes_franca_cidades.index[1]} - US$ {exportacoes_franca_cidades.iloc[1]["US$ FOB"]:,.2f}')

exportacoes_franca_cidades['US$ FOB'] = exportacoes_franca_cidades['US$ FOB'].map(formatar_valor)
display(exportacoes_franca_cidades)
```
