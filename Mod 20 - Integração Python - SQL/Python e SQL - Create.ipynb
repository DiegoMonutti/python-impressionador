```python
dados_conexao = ('Driver={SQLite3 ODBC Driver};'
                 'Server=localhost;'
                 'Database=Banco de Dados\chinook.db;')

conexao = pyodbc.connect(dados_conexao)
cursor = conexao.cursor()
```

```python
cursor.execute('''
INSERT INTO albums (Title, ArtistId)
VALUES
("Favourite Worst Nightmare", 276)
''')

cursor.commit()

cursor.close()
```
